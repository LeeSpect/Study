```python
### <- 이전 수업에 이어서
# 저번 영상에서 데이터를 준비해놨는데 본격적으로 선형 회귀를 하기에 앞서 해야 되는 것이 하나 더 있습니다.
# 모델의 성능을 신빙성 있게 평가하기 위해서 모델을 학습시키는 데이터와 모델을 평가하는 데이터를 나눈다고 얘기했었잖아요.
# 모델을 학습시키는 데이터를 트레이닝 셋이라고 하고 모델을 평가하는 데이터를 테스트 셋이라고 합니다.
# 이제 데이터를 트레이닝 셋과 테스트 셋으로 나눠봅시다. 다시 위로 올라가서 from sklearn.model_selection import train_test_split을 쓰고 다시 실행하겠습니다.
# 이 train_test_split 함수가 데이터를 트레이닝 셋과 테스트 셋으로 나눠주는데요. 일단 코드를 쓰고 설명하겠습니다.

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
boston_dataset = load_boston()

print(boston_dataset.DESCR)
```

```python
x = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
x
```

```python
x = x[['AGE']]

y = pd.DataFrame(boston_dataset.target, columns=['MEDV'])
y
```

```python
# 여기 train_test_split 함수의 파라미터로 입력 변수와 목표 변수를 모두 넘겨줬고요. test_size라는 파라미터로 0.2라는 값을 넘겨줬는데요.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)

# test_size는 전체 데이터 중에 20%만 골라서 테스트 셋으로 사용하고 나머지 80%는 트레이닝 셋으로 사용하라는 겁니다.
# 그리고 마지막에 random_state는 그 20%를 어떻게 고를지 정하는 파라미터입니다. 이건 옵셔널 파라미터이기 때문에 안 넘겨줘도 됩니다.
# 안 넘겨 주면 실행할 때마다 매번 랜덤한 20%를 테스트 셋으로 새롭게 고르게 됩니다.
# 그런데 지금처럼 어떤 정숫값을 넘겨주면 매번 20%의 똑같은 데이터를 고르게 됩니다.
# 일단 우리는 매번 똑같은 결과가 나오게 하기 위해서 정숫값 5를 넘겨주겠습니다.
# 아무 정수나 써도 되는데 일단 5를 쓰겠습니다.
# 이렇게 쓰면 train_test_split 함수는 4개의 값을 튜플로 리턴하는데요. 우리는 그 4개 값을 x_train, x_test, y_train, y_test. 총 4개의 변수에 저장한 겁니다.
# 트레이닝 셋의입력 변수는 x_train에 들어가고 테스트 셋의 입력 변수는 x_test에 들어가고 트레이닝 셋의 목표 변수는 y_train에 들어가고 테스트 셋의 목표 변수는 y_test에 들어가는 거죠.
```

```python
# 얘네들은 다 pandas 데이터 프레임인데요. 각각 어떤 모양인지 출력해보겠습니다.
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

# 보면 트레이닝 셋에는 404개의 데이터가 있고 테스트 셋에는 102개의 데이터가 들어간 걸 볼 수 있습니다.
# 우리가 여기에 테스트 셋의 크기를 20%로 정했으니까 20대 80으로 나뉜 거죠.
```
